```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense

from statsmodels.tsa.seasonal import STL

import math
```

```python
dataframe = pd.read_csv('output2.csv',usecols=[1]).iloc[:8000,:].clip(lower = 0)
dataframe
```

```python
plt.plot(dataframe)
```

```python
dataset = dataframe.values
dataset = dataset.astype('float32')
dataset.shape
```

# Preprocessing 

```python
scaler = MinMaxScaler(feature_range=(0,1))
dataset = scaler.fit_transform(dataframe)
```

```python
train_size = int(len(dataset)*0.80)
test_size = len(dataset) - train_size
```

```python
traindata = dataset[:train_size,:]
testdata = dataset[train_size:,:]
testdata.shape
```

```python
time_window , forecasting_term = 10,20
```

```python
def to_sequences(dataset,time_window,forecasting_term):
    X = []
    y= []
    
    for i in range(len(dataset) - time_window - forecasting_term):
        X.append(dataset[i:i + time_window,:].reshape(-1,1))
        y.append(dataset[i + time_window:i+ time_window + forecasting_term,:].reshape(-1,1))
        
    return np.array(X),np.array(y)
```

```python
trainX, trainY = to_sequences(traindata,time_window,forecasting_term)
trainY.shape
```

# CRUcell 

```python
hidden=2048
```

```python
wx_t = Dense(hidden*3, input_shape=(time_window,1))
wx_s = Dense(hidden*3, input_shape=(time_window,1))
wx_r = Dense(hidden, input_shape=(time_window,1))
```

```python
wh_t = Dense(hidden*3, input_shape=(hidden,1))
wh_s = Dense(hidden*3, input_shape=(hidden,1))
wh_r = Dense(hidden, input_shape=(hidden,1))
```

```python
hid_state = np.zeros((3, len(traindata)-time_window-forecasting_term, hidden) )
hid_state[0,:,:].shape
```

```python
output = []
```

```python
output = []
for i in range(time_window):
  
    lamda = 0.5
    x = pd.DataFrame(trainX[ : ,i,: ])
    dates = pd.date_range('1990-01-01', periods = len(x), freq='D')
    x.index = dates
    trend = []
    seasonal = []
    resid = []
    for i in range(len(x.columns)):
        stl = STL(x[i])
        res = stl.fit()
        trend.append(res.trend.values)
        seasonal.append(res.seasonal.values)
        resid.append(res.resid.values)
        
    trend = np.array(trend)
    seasonal = np.array(seasonal)
    resid = np.array(resid)
    
    x_t = tf.transpose(trend, perm=[1, 0])
    x_s = tf.transpose(seasonal, perm=[1, 0])
    x_r = tf.transpose(resid, perm=[1, 0])
        
    x_t = wx_t(x_t)
    x_s = wx_s(x_s)
    x_r = wx_r(x_r)
    
    h_t = wh_t(hid_state[0,:,:])
    h_s = wh_s(hid_state[1,:,:])
    
    x_autocor_t, x_cor_t, x_new_t = tf.split(x_t, num_or_size_splits=3, axis=1)
    x_autocor_s, x_cor_s, x_new_s = tf.split(x_s, num_or_size_splits=3, axis=1)
    h_autocor_t, h_cor_t, h_new_t = tf.split(h_t, num_or_size_splits=3, axis=1)
    h_autocor_s, h_cor_s, h_new_s = tf.split(h_s, num_or_size_splits=3, axis=1)
    
    autocor_t = tf.sigmoid(x_autocor_t + h_autocor_t)
    autocor_s = tf.sigmoid(x_autocor_s + h_autocor_s)
    
    cor_t = tf.sigmoid(x_cor_t + h_cor_s)
    cor_s = tf.sigmoid(x_cor_s + h_cor_t)
    
    new_t = lamda*tf.tanh(x_new_t + (autocor_t * h_new_t)) + (1-lamda)*tf.tanh(x_new_t + (cor_t * h_new_s))
    new_s = lamda*tf.tanh(x_new_s + (autocor_s * h_new_s)) + (1-lamda)*tf.tanh(x_new_s + (cor_s * h_new_t))
    
    hid_state[0,:,:] = new_t
    hid_state[1,:,:] = new_s
    hid_state[2,:,:] = tf.tanh(x_r)
    
    output.append( new_t + new_s + tf.tanh(x_r))
```

```python
fc = Sequential([
    tf.keras.layers.Flatten(input_shape=(hidden,)),  # Flatten the input
    tf.keras.layers.Dense(forecasting_term,activation='elu'),  # Dense layer
])
```
